```python
# load libraries
import pandas as pd
from matplotlib import pyplot as plt
import os
import numpy as np
from statsmodels.tsa.stattools import acf, pacf, adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from datetime import datetime as dt
from dateutil.relativedelta import relativedelta

plt.style.use('Solarize_Light2')
```

```python
# load data
items = pd.read_csv("./data/items.csv")
item_cat = pd.read_csv("./data/item_categories.csv")
sales = pd.read_csv("./data/sales_train.csv", 
                    parse_dates = [0], 
                    date_parser = lambda x: dt.strptime(x, "%d.%m.%Y"))
sample_sub = pd.read_csv("./data/sample_submission.csv")
shops = pd.read_csv("./data/shops.csv")
test = pd.read_csv("./data/test.csv")

sales
```

```python
# preprocessing
sales['net'] = sales.item_price * sales.item_cnt_day
sales = sales.join(items.set_index('item_id').item_category_id, on='item_id')
sales['date_m'] = sales.date.apply(lambda x: x.replace(day=1))
sales.rename({'item_cnt_day': 'item_cnt'}, axis=1, inplace=True)
```

```python
# records of negative item count in a day
sales.loc[sales.item_cnt <= 0].item_cnt.value_counts()
```

```python
agg_dict = {'item_price': ['mean','std','median','min','max','count'],
            'item_cnt': ['sum'],
            'net': ['sum'],
            'item_category_id': ['mean']}
total_sales_m = sales.groupby(['date_m']).agg(agg_dict).drop('item_category_id', axis=1)
item_sales_m = sales.groupby(['date_m', 'item_id']).agg(agg_dict)
shop_sales_m = sales.groupby(['date_m', 'shop_id']).agg(agg_dict).drop('item_category_id', axis=1)
item_shop_sales_m = sales.groupby(['date_m', 'shop_id', 'item_id']).agg(agg_dict)
item_cat_sales_m = sales.groupby(['date_m', 'item_category_id']).agg(agg_dict.pop('item_category_id'))
```

```python
total_sales_m[[('item_price','mean'), ('item_cnt','sum'),('net','sum')]].plot(subplots=True, figsize=(10,7))
```

## Total sales modeling

```python
tots_ts = total_sales_m[[('item_cnt','sum')]]
tots_ts.columns = ['item_cnt']
tots_ts = tots_ts.asfreq(pd.infer_freq(tots_ts.index))
tots_tsd = tots_ts.diff()[1:]
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,7))
tots_ts.plot(title="item count", ax=ax1)
tots_tsd.plot(title="first difference", ax=ax2)
ax2.axhline(0, linestyle='--', color='black', alpha=0.2)
plt.show()
```

```python
print("ADF p-value for first difference of total sales is", adfuller(tots_tsd)[1])
```

```python
nlags = 13
acf_vals = acf(tots_tsd, nlags)
pacf_vals = pacf(tots_tsd, nlags)
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,7))
#ax1.bar(range(nlags), acf_vals[:nlags])
plot_acf(tots_tsd, ax=ax1)
ax2.set_title('Partial Autocorrelation')
ax2.bar(range(nlags), pacf_vals[:nlags])
```

```python
# train on 2013-2014
train_data = tots_ts[:dt(2015,2,1)]
# test on 2015
test_data = tots_ts[dt(2015,3,1):]
```

```python
import pmdarima as pm
from sklearn.metrics import mean_squared_error
model = pm.auto_arima(train_data,error_action="ignore", suppress_warnings = True,
                         seasonal = True,
                         m = 12, D=0,
                         test='adf', #use adf test
                         information_criterion='aic', #AIC or BIC
                         stepwise = False, trace = False)
```

```python
model.summary()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

myorder = (0,1,0)
sorder = (0,1,0,12)
model = SARIMAX(train_data, order=myorder, seasonal_order=sorder)
model_fit = model.fit(suppress_warnings = True)
print(model_fit.summary())
```

```python
plt.plot(tots_ts)
plt.plot(model_fit.forecast(10))
```

```python
rolling_predictions = []
for train_end in test_data.index:
    train_data = tots_ts[:train_end - relativedelta(months=1)]    
    model = SARIMAX(train_data, order=myorder, seasonal_order=sorder)
    model_fit = model.fit(suppress_warnings = True)

    pred = model_fit.forecast()
    rolling_predictions.append(pred.values[0])
    # rolling_predictions[train_end] = pred
```

```python
plt.plot(tots_ts)
plt.plot(pd.DataFrame(rolling_predictions, index = test_data.index))
```
